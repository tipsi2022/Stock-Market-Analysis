```python
import talib as ta
import numpy as np
import pandas as pd
import yfinance as yf
from talib import MA_Type
from datetime import date
from talib import abstract
```

## DOWNLOAD HISTORICAL STOCK DATA FOR INFOSYS

```python
ticker = 'INFY'
start_time = '2021-01-02'
today = date.today()
end_time = today.strftime("%Y-%m-%d")
```

```python
df = pd.DataFrame()
df = yf.download(ticker, start = start_time, end = end_time)

df.head()
```

```python
df =  df.rename(columns={'Open': 'open', 'High': 'high', 'Low': 'low', 'Adj Close': 'close', 'Volume': 'volume'})
df = df.drop(['Close'], axis='columns')
df
```

## CALCULATE MOVING AVERAGES

```python
df['SMA'] = abstract.SMA(df, 20)
df['WMA'] = abstract.WMA(df, 20)
df['EMA'] = abstract.EMA(df, 20)
df['DEMA'] = abstract.DEMA(df, 20)
df['TEMA'] = abstract.TEMA(df, 20)

df
```

## CALCULATE MOMENTUM INDICATORS

```python
df['RSI'] = abstract.RSI(df, 14)

temp = abstract.MACD(df, fastperiod=12, slowperiod=26)
df['MACD'] = temp['macd']
df['MACD Signal'] = temp['macdsignal']

df['ADX'] = abstract.ADX(df, 14)
df['DI+'] = abstract.PLUS_DI(df, 14)
df['DI-'] = abstract.MINUS_DI(df, 14)

temp = abstract.STOCHF(df, fastk_period=14)
df['Stochastic Oscillator'] = temp['fastk']

df['Williams %R'] = abstract.WILLR(df, 14)

df
```

## EXTRACT CLOSING PRICE OF OTHER TICKERS

```python
temp = yf.download('TCS', start = start_time, end = end_time)
df['TCS Close'] = temp['Adj Close']
```

```python
df
```

```python
df =  df.rename(columns={'open': 'Open', 'high': 'High', 'low': 'Low', 'close': 'Target', 'volume': 'Volume'})
df
```

## CALCULATE RETURNS OF TARGET VARIABLE AND OTHER TICKERS

```python
df['Returns Target'] = (df['Target']/df['Target'].shift(1)) - 1
```

```python
df['Returns TCS'] = (df['TCS Close']/df['TCS Close'].shift(1)) - 1
```

```python
df
```

## CALCULATE VOLUME CHANGE

```python
df['Volume Change'] = df['Volume']/df['Volume'].shift(1)
```
