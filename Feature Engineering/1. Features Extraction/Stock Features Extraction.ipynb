```python
import talib as ta
import numpy as np
import pandas as pd
import yfinance as yf
from talib import MA_Type
from datetime import date
from talib import abstract
```

## DOWNLOAD HISTORICAL STOCK DATA FOR INFOSYS

```python
ticker = 'INFY'
start_time = '2019-12-04'
today = date.today()
end_time = today.strftime("%Y-%m-%d")
```

```python
df = pd.DataFrame()
df = yf.download(ticker, start = start_time, end = end_time)

df.head(30)
```

```python
df =  df.rename(columns={'Open': 'open', 'High': 'high', 'Low': 'low', 'Adj Close': 'close', 'Volume': 'volume'})
df = df.drop(['Close'], axis='columns')
df
```

## CALCULATION OF BOLLINGER BANDS

```python
temp = abstract.BBANDS(df, matype=0)

df['upper_band'] = temp['upperband']
df['lower_band'] = temp['lowerband']

df
```

## CALCULATION MOVING AVERAGES

```python
df['SMA_5'] = abstract.SMA(df, 5)
df['SMA_10'] = abstract.SMA(df, 10)
df['SMA_15'] = abstract.SMA(df, 15)
df['SMA_20'] = abstract.SMA(df, 20)

df['WMA_5'] = abstract.WMA(df, 5)
df['WMA_10'] = abstract.WMA(df, 10)
df['WMA_15'] = abstract.WMA(df, 15)
df['WMA_20'] = abstract.WMA(df, 20)

df['EMA_5'] = abstract.EMA(df, 5)
df['EMA_10'] = abstract.EMA(df, 10)
df['EMA_15'] = abstract.EMA(df, 15)
df['EMA_20'] = abstract.EMA(df, 20)

df['DEMA_5'] = abstract.DEMA(df, 5)
df['DEMA_10'] = abstract.DEMA(df, 10)
df['DEMA_15'] = abstract.DEMA(df, 15)
df['DEMA_20'] = abstract.DEMA(df, 20)

df['TEMA_5'] = abstract.TEMA(df, 5)
df['TEMA_10'] = abstract.TEMA(df, 10)
df['TEMA_15'] = abstract.TEMA(df, 15)
df['TEMA_20'] = abstract.TEMA(df, 20)

df
```

## CALCULATION OF MOMENTUM INDICATORS

```python
df['RSI'] = abstract.RSI(df, 14)

temp = abstract.MACD(df, fastperiod=12, slowperiod=26)
df['MACD'] = temp['macd']
df['MACD Signal'] = temp['macdsignal']

df['ADX'] = abstract.ADX(df, 14)
df['DI+'] = abstract.PLUS_DI(df, 14)
df['DI-'] = abstract.MINUS_DI(df, 14)

temp = abstract.STOCHF(df, fastk_period=14)
df['Stochastic Oscillator'] = temp['fastk']

df['Williams %R'] = abstract.WILLR(df, 14)

df
```

```python
df =  df.rename(columns={'open': 'Open', 'high': 'High', 'low': 'Low', 'close': 'Close', 'volume': 'Volume'})
df
```

## CALCULATION OF LAGGED CLOSING PRICES OF TARGET VARIABLE

```python
df['close_1'] = df['Close'].shift(1)
df['close_2'] = df['Close'].shift(2)
df['close_5'] = df['Close'].shift(5)

df
```

## CALCULATION OF DIFFERENCE OF CLOSING PRICE wrt PAST

```python
df['diff_1'] = df['Close'] - df['Close'].shift(1)
df['diff_2'] = df['Close'] - df['Close'].shift(2)
df['diff_5'] = df['Close'] - df['Close'].shift(5)

df
```

## CALCULATION OF RETURNS 

```python
df['returns_1'] = (df['Close']/df['Close'].shift(1)) - 1
df['returns_2'] = (df['Close']/df['Close'].shift(2)) - 1
df['returns_5'] = (df['Close']/df['Close'].shift(5)) - 1

df
```

## CALCULATION OF VOLUME CHANGE

```python
df['volume_change_1'] = (df['Volume']/df['Volume'].shift(1)) - 1
df['volume_change_2'] = (df['Volume']/df['Volume'].shift(2)) - 1
df['volume_change_5'] = (df['Volume']/df['Volume'].shift(5)) - 1

df
```

## EXTRACT CLOSING PRICES OF TCS

```python
temp = yf.download('TCS', start = start_time, end = end_time)
df['close_TCS'] = temp['Adj Close']

df['returns_TCS'] = (df['close_TCS']/df['close_TCS'].shift(1)) - 1

df
```

## REMOVING ROWS HAVING NaN VALUES

```python
df = df.dropna()
df
```
